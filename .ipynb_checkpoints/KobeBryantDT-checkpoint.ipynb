## Decision Tree Classifier using some features

#### Step 1. Reading data into the dataframe

```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)

original_df = pd.read_csv('data.csv')
print('Number of original rows: ', len(original_df.index))

pd.set_option('display.max_columns', None)
display(original_df)

# Shows the number of rows that are null for each col
display(original_df.isnull().sum())

# Drops all of the rows that are null
kobe_df = pd.read_csv('data.csv').dropna()
print('Number of rows after dropping: ', len(kobe_df.index))
```

## Show the data set

```python
# Show every 2000th entry of the data set.
display(kobe_df[0::2000])
```

## Since we have categorical data, we want to use One Hot Encoding to turn it into numerical data.

#### We get all of the features we are using first, and put them all into their own data sets.

```python
# Start off by excluding irrelevant features
unused_features = ['team_name', 'team_id', 'game_event_id','matchup' ,'game_id', 'shot_id', 'game_date', 'shot_made_flag']

# remove unused features from feature set (only has 16 useful categorical AND numerical features)
used_features = [feature for feature in kobe_df.columns if feature not in unused_features]

categorical_features = ['action_type', 'combined_shot_type','season','game_date', 'shot_type', 'shot_zone_area', 'shot_zone_basic', 'shot_zone_range', 'opponent']

numerical_features = ['lat', 'lon', 'loc_x', 'loc_y', 'minutes_remaining', 'period', 'playoffs', 'seconds_remaining', 'shot_distance']

# Our Label
target = ['shot_made_flag']

# Apply One-Hot Encoding
df_processed = pd.get_dummies(kobe_df[categorical_features])

# Show the data
df_processed.head(10)
```

## Now, let's split our dataset into testing and training sets with test_size = 0.3, random_state = 6

```python
# We will import train-test_split from sklearn's model_selection to split the data sets we will be using.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_processed, kobe_df[target].values.ravel(), test_size=0.30, random_state=6)
```

## After splitting, we can implement our Decision Tree Classifier, then test our accuracy.

```python
# Import Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

# Import accuracy score from sklearn.metrics
# We can use this to compare the predicted labels for the testing set with its actual labels.
from sklearn.metrics import accuracy_score

# Create instance of Decision Tree Classifier
decisionTree = DecisionTreeClassifier(random_state=5)

# Train on training set
decisionTree.fit(X_train, y_train)

# Test on testing set
labels_predict_dt = decisionTree.predict(X_test)

# Check accuracy
score_dt = accuracy_score(y_test, labels_predict_dt)

print("Accuracy of Decision Tree Classifier: ", score_dt)
```

## Using 10-fold Cross Validation

```python
# Import cross validation
from sklearn.model_selection import cross_val_score

accuracy_list = cross_val_score(decisionTree, df_processed, kobe_df[target].values.ravel(), cv=10, scoring='accuracy')

score_dt_cv = accuracy_list.mean()

print("Accuracy of Decision Tree Classifier using 10-Fold Cross Validation: ", score_dt_cv)
```

## The accuracy of a One-Hot Encoded Decision Tree Classifier with and without cross validation.

```python
print("Accuracy of Decision Tree Classifier with test_size = 0.30: ", score_dt)
print("Accuracy of Decision Tree Classifier using 10-Fold Cross Validation: ", score_dt_cv)
```

## Let's apply Ensemble Learning and use "Bagging" on 20 decision tree classifiers.

```python
from sklearn.utils import resample

# Create a list to store the prediction results
prediction_results = []

# Number of decision trees we want for voting.
num_dec_trees = 20

n_multiplier = 0.8

# Loop 20 times for i = 0 to i = 19. (20 decision tree classifiers)
for i in range(num_dec_trees):
    # Bootstrap the original with n_samples = 0.8 * size of original dataset
    bootstrap_train_X = resample(X_train, n_samples=(int)(n_multiplier*len(X_train)), random_state=i, replace=True)
    # Do the same with the label
    bootstrap_train_Y = resample(y_train, n_samples=(int)(n_multiplier*len(y_train)), random_state=i, replace=True)
    
    # Create a new decision tree classifier with random_state = 3
    Base_DecisionTree = DecisionTreeClassifier(random_state=3)
    
    # Train the new decision tree with the bootstrap sample
    Base_DecisionTree.fit(bootstrap_train_X, bootstrap_train_Y)
    
    # Predict with the old testing set
    bootstrap_predict = Base_DecisionTree.predict(X_test)
    
    # Store prediction results in list
    prediction_results.append(bootstrap_predict)
    
    # Showing the first 10 and last 5 results for every ith prediction
    print('i =',i,'Prediction:',bootstrap_predict[0:10], '...',bootstrap_predict[len(bootstrap_predict)-5:-1])
    
```

### With the predictions, we can perform voting to make the final decision

```python
# Using math.ceil()
from math import ceil

# Create a list to store our final predictions
final_prediction = []

# Ceil rounds a float to the highest integer
majority = ceil(num_dec_trees/2)
ct = 0

# Sum of columns for each row
for i in range(len(prediction_results[0])):
    # Add whatever is in the column in that row to the count
    for j in range(num_dec_trees):
        ct += prediction_results[j][i]
        
    # Once everything is added, check if the count is < majority and append the result to the final_prediction
    if (ct < majority):
        final_prediction.append(0)
    else:
        final_prediction.append(1)
    ct = 0

# Check accuracy
accuracy_dt_bagging = accuracy_score(y_test, final_prediction)

print('Accuracy of using Bagging: ', accuracy_dt_bagging)
```

## Now using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Using random forest classifier
my_RandomForest = RandomForestClassifier(n_estimators = 200, bootstrap=True, random_state=3)
my_RandomForest.fit(X_train, y_train)
rf_predict = my_RandomForest.predict(X_test)

print('Prediction of Random Forest Classifier: ', rf_predict)

accuracy_dt_rf = accuracy_score(y_test, rf_predict)
print('Accuracy of using Random Forest Classifier: ', accuracy_dt_rf)
```

```python
print("Accuracy of Decision Tree Classifier with test_size = 0.30: ", score_dt)
print("Accuracy of Decision Tree Classifier using 10-Fold Cross Validation: ", score_dt_cv)
print('Accuracy of using Bagging: ', accuracy_dt_bagging)
print('Accuracy of using Random Forest Classifier: ', accuracy_dt_rf)
```
